```python
def same_names(users_input):
    
    '''If user's query gives 2 or 3 results, function prints a list of 2 or 3 artists (without repetitions), 
        asks user which one they were looking for and returns the track_id of the chosen one.'''
        
    title=users_input
    
    title_1 = sp.search(q=title,limit=3,market="GB")["tracks"]["items"][0]["name"]
    title_2 = sp.search(q=title,limit=3,market="GB")["tracks"]["items"][1]["name"]
    title_3 = sp.search(q=title,limit=3,market="GB")["tracks"]["items"][2]["name"]
    
    artist_1 = sp.search(q=users_input,limit=3,market="GB")["tracks"]["items"][0]["artists"][0]["name"]
    artist_2 = sp.search(q=users_input,limit=3,market="GB")["tracks"]["items"][1]["artists"][0]["name"]
    artist_3 = sp.search(q=users_input,limit=3,market="GB")["tracks"]["items"][2]["artists"][0]["name"]        
       
    
    id_1 = sp.search(q=title,limit=3,market="GB")["tracks"]["items"][0]["id"]
    id_2 = sp.search(q=title,limit=3,market="GB")["tracks"]["items"][1]["id"]
    id_3 = sp.search(q=title,limit=3,market="GB")["tracks"]["items"][2]["id"] 
  
    artists_list = [artist_1, artist_2, artist_3]
    ids_list = [id_1, id_2, id_3]    
    artist_id_dict = dict(zip(artists_list, ids_list)) 

    
    if len(list(artist_id_dict.keys())) == 1:
        key1 = (list(artist_id_dict.keys())[0])
        users_track_id = artist_id_dict[key1]
    
        return users_track_id
        
    elif len(list(artist_id_dict.keys())) == 2:
        key1 = (list(artist_id_dict.keys())[0])
        key2 = (list(artist_id_dict.keys())[1])
        
        print("For " + str(title_1) + " by " + "\x1b[1;31m" + artist_1 + '\x1b[0m' + " >>>> press 1")
        print("For " + str(title_2) + " by " + "\x1b[1;31m" + artist_2 + '\x1b[0m' + " >>>> press 2")
        answer = input("Please enter a number ")
        
        if answer == "1":
            users_track_id = artist_id_dict[key1]
            return users_track_id
        elif answer == "2":
            users_track_id = artist_id_dict[key2]
            return users_track_id
        else:
            print("You missed the button, try again :)")    
        
        
    elif len(list(artist_id_dict.keys())) == 3:
        key1 = (list(artist_id_dict.keys())[0])
        key2 = (list(artist_id_dict.keys())[1])
        key3 = (list(artist_id_dict.keys())[2]) 
        
        print("For " + str(title_1) + " by " + "\x1b[1;31m" + artist_1 + '\x1b[0m' + " >>>> press 1")
        print("For " + str(title_2) + " by " + "\x1b[1;31m" + artist_2 + '\x1b[0m' + " >>>> press 2")
        print("For " + str(title_3) + " by " + "\x1b[1;31m" + artist_3 + '\x1b[0m' + " >>>> press 3")
        answer = input("Please enter a number ")
       
        if answer == "1":
            users_track_id = artist_id_dict[key1]
            return users_track_id
        elif answer == "2":
            users_track_id = artist_id_dict[key2]
            return users_track_id
        elif answer == "3":
            users_track_id = artist_id_dict[key3]
            return users_track_id
        else: 
            print("You missed the button, try again :)")      

     
# users_track_id = same_names(users_input)
# print(users_track_id)
```

```python
def rec_from_id(users_track_id):
    
    '''Function takes final_id from the same_names() function as input and then: 
        1.gets audio features from spotipy and converts them into dataframe, 
        2. scales them with the standard scaler (output: np.array), 
        3. converts them to dataframe again, 
        4. uses kmeans from pickle to find the corresponding cluster in the song_df, 
        5. picks random row with the same cluster, 
        6. extracts title and artist of the recommendation and prints it, 
        7. returns track_id of the recommended song'''    
    
    users_track_features = pd.DataFrame(sp.audio_features(users_track_id)) # getting features from id
    
    users_track_features.drop(users_track_features[['type', 'uri','track_href','analysis_url','duration_ms','id','time_signature','mode']], axis = 1, inplace=True)
     
    users_track_scaled = scaler.transform(users_track_features) # scaling, output: np.array
    
    users_track_scaled_df = pd.DataFrame(users_track_scaled) # putting np.array with scaled features into df
    
    users_track_cluster = kmeans.predict(users_track_scaled_df) # finding the cluster
    
    recommendation = our_db[our_db["cluster"] == int(users_track_cluster)].sample() # random row for recommendation
    
    recommendation_id = recommendation.iloc[0]['id'] # getting id from the recommendation row
    
    print("Recommended song: " + "\x1b[1;31m" + str(recommendation.iloc[0]["title"]) + '\x1b[0m' + " by " + str((recommendation.iloc[0]["artist"]))) # printint recommendation
   
    return recommendation_id

#print(rec_from_id(users_track_id))
```
